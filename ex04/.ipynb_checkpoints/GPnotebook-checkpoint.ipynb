# Gaussian processes

Disclaimer: All examples in this notebook are for univariate x, and provided solely for understanding the concept rather than as general-purpose library for working with GPs. GPs work directly for multivariate x without any changes in equations, but the univariate case helps avoiding more complicated code for handling the dimensionalities. The kind of plots we use for visualizing the functions are only possible for the univariate case.

```python
import matplotlib.pyplot as plt
import torch
import numpy as np
import os
import numpy as np
import scipy.stats as stats
%matplotlib inline
```

```python
# Sample functions from GP prior

def meanFunc(x):
    return 0

def covFunc(a, b, l=1.):
    return np.exp(-0.5*(a-b)**2/l)

def covMatrix(a, b, l=1., lam=1.):
    # You usually would not compute this with nested for loops, but
    # I wanted to make it very clear what the kernel looks like
    k = np.zeros((len(a), len(b)))
    for i in range(len(a)):
        for j in range(len(b)):
            k[i,j] = covFunc(a[i],b[j], l=l)
    return lam*k
    
# Set of inputs to evaluate the function for
x = np.arange(-5,5,0.1)
print(x.shape)
m = [meanFunc(a) for a in x]
k = covMatrix(x, x, l=4.)
# A tiny bit of extra for diagonal, to make sure k is positive definite for sampling
k = k + np.eye(len(x))*1e-7

plt.rcParams["figure.figsize"] = (14, 8)
# Inspect the kernel
plt.subplot(2,2,1)
plt.title("Kernel function")
xsub = np.linspace(-10.,10.,50)
plt.plot(xsub, covFunc(xsub,0,l=.5))
plt.subplot(2,2,2)
plt.imshow(k)
plt.title("Covariance for linearly spaced inputs")

# Draw some samples -- for any given set of coordinates to evaluate this we
# simply draw the values from multivariate normal distribution
plt.subplot(2,2,3)
for n in range(20):
    y = stats.multivariate_normal(m, k).rvs(1)
    plt.plot(x, y, linewidth=0.25)
plt.title("Prior samples, smooth functions")

k = covMatrix(x, x, l=.3) + np.eye(len(x))*1e-7
plt.subplot(2,2,4)
for n in range(20):
    y = stats.multivariate_normal(m, k).rvs(1)
    plt.plot(x, y, linewidth=0.25)
plt.title("Prior samples, not so smooth functions")
```

```python
# Compute the posterior with analytic expressions
# Note that this is maximally naive implementation to match the slides
def posterior(y_obs, k_obs, k_pred, k_pred_obs):
    alpha = np.matmul(k_pred_obs, np.linalg.inv(k_obs))
    mu_pred = np.matmul(alpha, y_obs)
    cov_pred = k_pred - np.matmul(alpha, np.transpose(k_pred_obs))
    return mu_pred, cov_pred

# Sample one true function from the prior
k = covMatrix(x, x, l=.4) + np.eye(len(x))*1e-7
y = stats.multivariate_normal(m, k).rvs(1)

# Observe first 30 samples and last 30 samples
train = np.hstack([np.arange(30),np.arange(70,100)])
test = np.arange(30,70)

# Plot posterior functions compatible with the observed values, together with the mean estimate
mpred, kpred = posterior(y[train], k[train,:][:,train], k[test,:][:,test], k[test,:][:,train])
plt.plot(x[train],y[train],'.')
for n in range(20):
    y_pred = stats.multivariate_normal(mpred, kpred).rvs(1)
    plt.plot(x[test],y_pred,linewidth=0.25)
_ = plt.plot(x[test], mpred, 'k.')
```

```python
plt.rcParams["figure.figsize"] = (15, 40)

# Consecutively add more observations, this time with some observation noise too
noise = .4
l = .5

# True function
k = covMatrix(x,x,l=l) + np.diag(np.full(len(x), 1e-7))
y_mean = stats.multivariate_normal(m, k).rvs(1)

# Observe some data points and pre-sample noise for all of those
Nobs = 16
train = np.random.choice(np.arange(len(x)),Nobs)
y_noise = np.random.randn(Nobs)*noise

k_pred = covMatrix(x, x, l=l)
for nt in range(0,len(train)):
    plt.subplot(8,2,nt+1)
    if nt>0:
        k_obs = covMatrix(x[train[:nt]], x[train[:nt]], l=l)
        # Observation noise handled by adding it on the diagonal
        k_obs += np.diag(np.full(nt, noise**2.0))
        k_pred_obs = covMatrix(x, x[train[:nt]], l=l)
        y_obs = y_mean[train[:nt]] + y_noise[:nt]
        mpred, covpred = posterior(y_obs, k_obs, k_pred, k_pred_obs)
        if nt == 10:
            print(x.shape)
            print(x[train[:nt]].shape)
            print(x.shape)
            print(k_obs.shape)
            print(k_pred_obs.shape)
            print(mpred)
            print(covpred.shape)
    else:
        # Before any data we just have the prior
        # You could formally use posterior() for this with empty k_obs etc, but
        # would need to write it properly to suppor that
        mpred = np.zeros(len(x))
        covpred = np.diag(np.ones(len(x)))
    
    plt.plot(x, mpred, 'k-')
    if nt>0:
        plt.plot(x[train[:nt]], y_obs, 'r*')
    plt.plot(x, mpred + 2*np.diag(covpred), 'k:')
    plt.plot(x, mpred - 2*np.diag(covpred), 'k:')
    plt.plot(x, y_mean, 'b--')
```
