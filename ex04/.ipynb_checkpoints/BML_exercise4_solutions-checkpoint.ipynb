# Problem 1: Hyperparameter optimization

For selecting the hyperparameters we can use almost any algorithm that maximizes the marginal likelihood. Below we have simple gradient-based optimization routine, parameterizing all three parameters with their logarithms since they all need to be positive. For just 1 or 2 parameters I would recommend using a simple grid search instead, but for 3 it already starts to be a bit cumbersome because setting the grid with adequate steps is not trivial.

Note that during this optimization loop we do not need to compute the posterior as such. We are optimizing for the marginal likelihood and it is enough to evaluate that for the different choices.

```python
import torch 
import matplotlib.pyplot as plt

torch.manual_seed(7879)

def covFunc(a, b, l=1.):
    return torch.exp(-0.5*(a-b)**2./l**2.0)

def covMatrix(a, b, l=1.):
    # You usually would not compute this with nested for loops, but
    # I wanted to make it very clear what the kernel looks like
    k = torch.zeros((len(a), len(b)))
    for i in range(len(a)):
        for j in range(len(b)):
            k[i,j] = covFunc(a[i],b[j], l=l)
    return k
    
# Compute the posterior with analytic expressions
def posterior(y_obs, k_obs, k_pred, k_pred_obs):
    invmat = torch.inverse(k_obs)
    mu_pred = torch.matmul( torch.matmul(k_pred_obs, invmat), y_obs )
    cov_pred = k_pred - torch.matmul( torch.matmul( k_pred_obs, invmat ) , torch.t(k_pred_obs) )
    return mu_pred, cov_pred

# Read in data
data = np.loadtxt("problem1data.csv", delimiter=',')
x = torch.from_numpy(data[:,0]).float()
y = torch.from_numpy(data[:,1]).float()
Nobs = 30

# Define as optimization problem, starting with 1 for all parameters
# Note that for some configurations you might converge to a different local optimum
log_noise = torch.tensor(0., requires_grad=True)
log_l = torch.tensor(0., requires_grad=True)
log_lam = torch.tensor(0., requires_grad=True)
optimizer = torch.optim.Adam([log_noise,log_l,log_lam], lr=0.01)

for i in range(500):
    k_obs = torch.exp(log_lam)*covMatrix(x, x, l=torch.exp(log_l)) + torch.eye(Nobs)*torch.exp(log_noise)**2.0

    # The objective is the marginal likelihood that tells how well the observed
    # data fits the model in general
    invmat = torch.inverse(k_obs)
    loss = 0.5*torch.logdet(k_obs) + 0.5*torch.matmul(torch.matmul(y, invmat), y) + 0.5*Nobs*np.log(2.*np.pi)
    
    if i%50==0:
        print(torch.exp(log_lam).item(), torch.exp(log_noise).item(), torch.exp(log_l).item(), -loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
print('Optimal lambda:',torch.exp(log_lam).item())
print('Optimal noise std:',torch.exp(log_noise).item())
print('Optimal length scale:',torch.exp(log_l).item())
print('Optimal marginal likelihood:',-loss.item())
```

Below we can see the true parameters I originally used for generating the data. We do not have perfect match here because we only have 30 observations and are estimating a point estimate for these hyperparameters, but the selected ones are very close to the ones that were used. Also note that the marginal likelihood for the true parameters is a bit worse too, as it should.

```python
k_obs = 1.5*covMatrix(x, x, l=np.sqrt(0.5)) + torch.eye(Nobs)*0.3**2.0
invmat = torch.inverse(k_obs)
loss = 0.5*torch.logdet(k_obs) + 0.5*torch.matmul(torch.matmul(y, invmat), y) + 0.5*Nobs*np.log(2.*np.pi)

print('True generating lambda:', 1.5)
print('True generating noise std:', 0.3)
print('True generating length scale:', np.sqrt(0.5))
print('Marginal likelihood for these parameters:', -loss)
```

Now we can inspect how the hyperparameters influence the solution. The code below visualizes the posterior using (a) the optimal parameters and (b) three other choices that all have clearly marginal likelihood.

```python
import numpy as np
os.environ['KMP_DUPLICATE_LIB_OK']='True'

plt.rcParams["figure.figsize"] = (14, 6)

x_pred = torch.linspace(-2.5, 2.5, 100)

lam = torch.exp(log_lam)
l = torch.exp(log_l)
noise = torch.exp(log_noise)

plt.subplot(2,2,1)
k_obs = lam*covMatrix(x, x, l=l) + torch.eye(Nobs)*noise**2.0
k_pred = lam*covMatrix(x_pred, x_pred, l=l)
k_pred_obs = lam*covMatrix(x_pred, x, l=l)

mpred, kpred = posterior(y, k_obs, k_pred, k_pred_obs)
mpred = mpred.detach().numpy()
kpred = kpred.detach().numpy()
plt.plot(x_pred, mpred, 'k-')
plt.title("Optimal parameters")
plt.plot(x, y, 'b*')
# Uncertainty over the function values
plt.plot(x_pred, mpred + 2*np.sqrt(np.diag(kpred)), 'k:')
plt.plot(x_pred, mpred - 2*np.sqrt(np.diag(kpred)), 'k:')
# Uncertainty for new observed data
plt.plot(x_pred, mpred + 2*np.sqrt(np.diag(kpred) + noise.detach().numpy()**2.0), 'r:')
plt.plot(x_pred, mpred - 2*np.sqrt(np.diag(kpred) + noise.detach().numpy()**2.0), 'r:')

# Three examples with bad values for each of the parameters
for bad in range(3):
    plt.subplot(2,2,2+bad)
    lamplot = 1.5
    noiseplot = 0.3
    lplot = np.sqrt(0.5)
    if bad==0:
        lamplot = 0.01
    if bad==1:
        noiseplot = 2.0
    if bad==2:
        lplot = 10.0
        
    k_obs = lamplot*covMatrix(x, x, l=lplot) + torch.eye(Nobs)*noiseplot**2.0
    k_pred = lamplot*covMatrix(x_pred, x_pred, l=lplot)
    k_pred_obs = lamplot*covMatrix(x_pred, x, l=lplot)
    
    mpred, kpred = posterior(y, k_obs, k_pred, k_pred_obs)
    mpred = mpred.detach().numpy()
    kpred = kpred.detach().numpy()
    plt.plot(x_pred, mpred, 'k-')
    if bad==0:
        plt.title("Small lambda")
    if bad==1:
        plt.title("Large noise")
    if bad==2:
        plt.title("Large length-scale")
    plt.plot(x, y, 'b*')
    # Uncertainty over the function values
    plt.plot(x_pred, mpred + 2*np.sqrt(np.diag(kpred)), 'k:')
    plt.plot(x_pred, mpred - 2*np.sqrt(np.diag(kpred)), 'k:')
    # Uncertainty for new observed data
    plt.plot(x_pred, mpred + 2*np.sqrt(np.diag(kpred) + noiseplot**2.0), 'r:')
    _ = plt.plot(x_pred, mpred - 2*np.sqrt(np.diag(kpred) + noiseplot**2.0), 'r:')

```

We see that with the optimal hyperparameters the fit for the data is very good and that we could reliably use this model for making predictions.

Too small $\lambda$ pulls the function towards the mean, too large $\sigma$ results in very wide confidence intervals and smoother function as it does not need to pass so close to the data, and too large $l$ makes the function way too smooth, converging approximately to linear regression.

The plots illustrate uncertainty in two ways: The black dotted lines provide 95% intervals for the values of the unknown function $f$, whereas the red ones do the same for predicted outputs $y$. It is important to note that for the very bad choices of hyperparameters these confidence intervals are not very sensible. For the optimal parameters the actual observations follow nicely the range of values we expect to see. The individual observations are nicely within the red confidence intervals as they should, but can well be outside of the black ones as they are only noisy observations of the underlying function.

# Problem 2: Poisson regression

A code implementing the Poisson regression is provided below. It follows exactly the basic principle of Laplace approximation by first finding the MAP estimate for the unknown function and the forming the Hessian around that point to estimate the covariance of the approximation.

The non-trivial parts of the solution are roughly:
1. To optimize for the function $f(x_i)$ we explicitly parameterize it as a vector of values $f_i$, one for each observed $x_i$. Then we can simply optimize over $f=[f_1,...,f_N]$.
2. The log-likelihood $p(y_i|f(x_i))$ for each observed data point is $y_i \log \lambda - \lambda = y_i f_i - e^{f_i} - \log y_i!$ since we model the rate parameter for a Poisson distribution with $e^{f_i}$. Note that the constant term $\log y_i!$ is ignored in the code as it does not matter for MAP at all.
3. The 2nd derivative of the likelihood, required for the variance of the Gaussian approximation, is obtained by
  $
  \frac{d \log p(y|f)}{df} = y - e^{f}$ and $\frac{d^2 \log p(y|f)}{d^2f} = -e^{f}$. Then the covariance for the Laplace approximation will be $(K^{-1} + W)^{-1}$ where $W_{ii} = - \frac{d^2 \log p(y_i|f_i)}{d^2f_i} = e^{f_i}$. By comparing this expression with the covariance of a conjugate model (that is, normal likelihood for observations) we see that the inverse of the diagonal elements of $W$ acts like the variance for the observations.
4. For plotting we need to remember that $E[e^f]$ does not equal $e^{E[f]}$. This means that we cannot first compute the mean and variance of $f$ and then transform them to the observation space, but instead we need to transform each posterior sample independently and then compute the mean and variance of those for visualization. The plots below do not compute variances but rather visualise the uncertainty directly by plotting draws from the posterior approximation.

```python
torch.manual_seed(76)

# Read in the data
data = np.loadtxt("problem2data.csv", delimiter=',')
x = torch.from_numpy(data[:,0]).float()
y = torch.from_numpy(data[:,1]).float()
Nobs = 40

# Hyperparameters as given in the exercise sheet
ltrue = .5
lamtrue = 2.

# Solve for the MAP
f = torch.randn((Nobs), requires_grad=True)
optimizer = torch.optim.Adam([f], lr=0.1)
# Add a small amount of 'noise' to improve computational stability
k_obs = lamtrue*covMatrix(x, x, l=ltrue) + torch.eye(Nobs)*1e-5
invmat = torch.inverse(k_obs)

for i in range(10000):
    loss = 0.5*torch.logdet(k_obs) + 0.5*torch.matmul(torch.matmul(f, invmat), f) - torch.sum(y*f - torch.exp(f))
    
    if i%1000==0:
        print(loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Compute variance as negative 2nd derivative
#logp = y*f - exp(f)
#dlogp = f - exp(f)
#d2logp = -exp(f)
diagW = torch.exp(f) 

# Posterior covariance approximation
S = torch.inverse(invmat + torch.diag(diagW))
mvn = torch.distributions.MultivariateNormal(f, covariance_matrix = S + torch.eye(Nobs)*1e-3)

Npred = 100
xpred = torch.linspace(-5, 5, Npred)
k_obs = lamtrue*covMatrix(x, x, l=ltrue) + torch.diag(1./diagW)
k_pred = lamtrue*covMatrix(xpred, xpred, l=ltrue)
k_pred_obs = lamtrue*covMatrix(xpred, x, l=ltrue)
pmean, pcov = posterior(f, k_obs, k_pred, k_pred_obs)
mvnpred = torch.distributions.MultivariateNormal(pmean, covariance_matrix = pcov + torch.eye(Npred)*1e-3)


plt.subplot(1,2,1)

# Predictions for arbitrary x in the log-space
plt.plot(xpred, pmean.detach(), 'r-')
for n in range(30):
    f2 = mvnpred.sample()
    plt.plot(xpred, f2, 'r-', linewidth=0.1)
# Observed data transformed to (roughly) the same space the functions live in
# Small constant is needed as log(0)=-inf
plt.plot(x, np.log(y+0.1), 'b*', markersize=3)
plt.title("Log-space")

    
plt.subplot(1,2,2)

# Prediction for arbitrary x in the real observation space
# Note that we cannot here use np.exp(pmean) as it would not correspond to the mean prediction
for n in range(30):
    f2 = mvnpred.sample()
    plt.plot(xpred, np.exp(f2), 'r-', linewidth=0.1)
plt.plot(x, y, 'b*', markersize=3)
_ = plt.title("Observation space")
```

The plots show that the estimated function quite nicely aligns with the observed data. For the region with no observed data the predictions rather quickly fall back towards zero in the middle -- this is because we are using GP mean function of zero and hence expected rate of $e^0=1$. Since e.g. $x=0$ is already four times the length-scale away from the closest data points, the correlation between $f(0)$ and all of the observed cases is essentially zero and hence the prior itself is the best estimate we can have.

After making these observations we might want to improve the model a bit. Looking at the left plot we see that for the observed data the true mean of the function is closer to 2. Maybe changing the prior to GP(2, k(x,x')) would make sense here. Then the mean predictions in the observed space would settle around $e^2 \approx 7$ rather than $e^0=1$.
