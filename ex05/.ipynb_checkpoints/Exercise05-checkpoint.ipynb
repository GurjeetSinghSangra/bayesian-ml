# Gradients for Variational Approximation

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import os
import torch
%matplotlib inline
plt.rcParams["figure.figsize"] = (12, 6)

x = torch.tensor(np.array([3, 4, 3, 9, 10, 3, 2, 3, 3, 2]))
```

## Monte Carlo estimator

```python
def elbo_monte_carlo_estimate(x, q_u, q_v, M):
    q_u_samples = q_u.sample((M, ))
    q_v_samples = q_v.sample((M, ))
    log_joint_samples = torch.zeros((M))
    for i in range(0, M):
        likelihood = torch.distributions.Poisson(q_u_samples[i]*q_v_samples[i]) 
        log_joint_samples[i] = torch.sum(likelihood.log_prob(x))
    return -torch.mean(log_joint_samples - q_u.log_prob(q_u_samples) - q_v.log_prob(q_v_samples))
```

```python
mu_u = torch.nn.Parameter(torch.tensor(1.))
sigma_u = torch.nn.Parameter(torch.tensor(0.1))
mu_v = torch.nn.Parameter(torch.tensor(2.))
sigma_v = torch.nn.Parameter(torch.tensor(.1))

optimizer = torch.optim.SGD({mu_u, sigma_u, mu_v, sigma_v}, lr=1e-4)
n_epochs = 1000
M = 50
loss_values = torch.zeros(n_epochs)

for i in range(n_epochs):
    q_u = torch.distributions.Normal(mu_u, sigma_u)
    q_v = torch.distributions.Normal(mu_v, sigma_v)
    loss = elbo_monte_carlo_estimate(x, q_u, q_v, M)
    loss_values[i] = loss.clone()
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
plt.plot(np.arange(0, n_epochs), loss_values.detach())

print(mu_u.detach())
print(sigma_u.detach())
print(mu_v.detach())
print(sigma_v.detach())
```

## Score function estimator

```python
def log_norm(mu, sigma, samples):
    dist = torch.distributions.Normal(mu, sigma)
    logprobs = dist.log_prob(samples)
    return logprobs

#mu_u = torch.nn.Parameter(torch.tensor(1.))
mu_u = torch.tensor(5.0)
sigma_u = torch.tensor(0.5)
mu_v = torch.tensor(1.)
sigma_v = torch.tensor(0.2)

lr = 1e-3
n_epochs = 50
loss_values = torch.zeros(n_epochs)
M = 1000
for epoch in range(n_epochs):
    q_u = torch.distributions.Normal(mu_u, sigma_u)
    q_v = torch.distributions.Normal(mu_v, sigma_v)
    
    q_u_samples = q_u.sample((M, 1))
    q_v_samples = q_v.sample((M, 1))
    
    W = torch.zeros((M, 1))
    for j in range(0, M):
        likelihood = torch.distributions.Poisson(q_u_samples[j]*q_v_samples[j]) 
        W[j] = torch.sum(likelihood.log_prob(x))
    
    W = W - q_u.log_prob(q_u_samples) - q_v.log_prob(q_v_samples)
    
    ###### u Update
    u_jacobian = torch.autograd.functional.jacobian(log_norm, (mu_u, sigma_u, q_u_samples))
    #Jacobian for mu and sigma for each samples
    u_jacobian = torch.hstack((u_jacobian[0], u_jacobian[1]))
    # Gradient of the Elbo for each component of u
    u_elbo_grad = (1/M * torch.matmul(u_jacobian.T, W))
    mu_u = mu_u - lr * u_elbo_grad[0].item()
    sigma_u = sigma_u - lr * u_elbo_grad[1].item()
    
    
    ###### v Update
    v_jacobian = torch.autograd.functional.jacobian(log_norm, (mu_v, sigma_v, q_v_samples))
    #Jacobian for mu and sigma for each samples
    v_jacobian = torch.hstack((v_jacobian[0], v_jacobian[1]))
    v_elbo_grad = (1/M * torch.matmul(v_jacobian.T, W))

    # Gradient of the Elbo for each component of u
    mu_v = mu_v - lr * v_elbo_grad[0].item()
    sigma_v = sigma_v - lr * v_elbo_grad[1].item()
    
```
