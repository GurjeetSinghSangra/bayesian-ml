# Gibbs Sampling

Gibbs sampling (from previous notebook) as reference, so that we can compare the results. Now we plot also marginal distribution of log-precision, to compare with Laplace approximation.

```python
import numpy as np

def Gibbs(x, mu, prec, T):
    # Storage for samples
    samples = [mu, prec]
    for t in range(T):
        # Sample mu conditional on precision
        S2 = len(x)*prec + 1.0/prior_sigma**2.0
        m = (prec*np.sum(x) + 1.0/prior_sigma**2.0 * prior_mu) / S2
        mu = np.random.normal(m, np.sqrt(1/S2))
        
        # Sample prec conditional on mu
        prec = np.random.gamma(prior_alpha + 0.5*len(x), scale = 1.0 / (prior_beta + 0.5*np.sum((x - mu)**2.0)))

        samples = np.vstack([samples, [mu, prec]])
    return samples

x = np.array([5.21081825, 4.81293801, 4.92248705, 5.17930311, 6.41654995, 4.30446395, 5.34353883, 4.21838874, 3.31724447, 5.71822463])
gibbs_samples = Gibbs(x, 3.0, 0.5, 5000)
```

# Demonstration of distributional approximation algorithms for the example model

First define the model, giving some reasonable priors. Note that we are using PyTorch this time, since we want to use its optimization functionality.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import torch
import torch.distributions as dist
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

# Define the same model as for MCMC
prior_mu = 0.0
prior_sigma = 3.0
prior_alpha = 10.0
prior_beta = 5.0
mean_prior = dist.normal.Normal(torch.tensor([prior_mu]), torch.tensor([prior_sigma]))
prec_prior = dist.gamma.Gamma(torch.tensor([prior_alpha]), torch.tensor([prior_beta]))

# Re-use same data
mu_true = 4.8289413
prec_true = 2.57470688
x = torch.tensor([5.21081825, 4.81293801, 4.92248705, 5.17930311, 6.41654995, 4.30446395, 5.34353883, 4.21838874, 3.31724447, 5.71822463])

def log_posterior(data, mu, prec):
    likelihood = dist.normal.Normal(mu, torch.sqrt(1.0/prec))
    return torch.sum(likelihood.log_prob(data)) + mean_prior.log_prob(mu) + prec_prior.log_prob(prec)
```

## Laplace approximation

Laplace approximation proceeds by first finding the mode of log-density, and then estimating the variance around that point. We use SGD for the former and automatic differentiation (with very ugly PyTorch code -- this is almost certainly now how it should be done) for estimating the curvature, even though we could also derive e.g. coordinate ascent for optimization and compute analytic second derivatives for the curvature.

Since precision needs to be positive, we again use logarithmic parameterization for that.

```python
# First find the best point estimate, parameterizing precision in log-domain
mu = torch.nn.Parameter(torch.randn(()))
log_prec = torch.nn.Parameter(torch.randn(()))

optimizer = torch.optim.SGD({mu, log_prec}, lr=1e-3)
for iter in range(2000):
    loss = -log_posterior(x, mu, torch.exp(log_prec))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
print("MAP estimate for mean: ", mu.item())
print("True mean: ", mu_true)
print("MAP estimate for precision: ", torch.exp(log_prec).item())
print("True precision: ", prec_true)

# PyTorch supports direct computation of the Hessian, but needs specific syntax
def lossForHessian(param):
    return - torch.sum(dist.normal.Normal(param[0], torch.sqrt(1.0/torch.exp(param[1]))).log_prob(x)) - mean_prior.log_prob(param[0]) - prec_prior.log_prob(torch.exp(param[1]))
param = torch.tensor([mu.item(), log_prec.item()])
H = torch.autograd.functional.hessian(lossForHessian, param)
print("Precision of the approximation: ", H)

# Form the actual approximation
laplace = dist.multivariate_normal.MultivariateNormal(torch.tensor([mu, log_prec]), precision_matrix=H)
laplace_mu = dist.normal.Normal(mu, scale=1.0/torch.sqrt(H[0,0]))
laplace_log_prec = dist.normal.Normal(log_prec, scale=1.0/torch.sqrt(H[1,1]))
```

Inspect results by looking at the marginal distribution, overlaid on top of the Gibbs results. We could also plot the joint posterior, but the marginals already give a good picture of what is happening.

We see that the marginal distributions very closely match the ones we got with MCMC, and there is no reason to expect the same would not be true for the joint. Note that the two variables are not independent (the off-diagonal elements in H are not zero), but almost are.

Another important observation is that both marginals are normal distributions, as was assumed. Note, however, that the distribution for precision itself would not be.

```python
# Marginal distribution for mu
ind = torch.tensor(np.arange(mu_true-1.5,mu_true+1.5,0.01),dtype=torch.double)
plt.hist(gibbs_samples[1000:,0], bins=20, density=True)
plt.plot(ind, torch.exp(laplace_mu.log_prob(ind)).detach())
plt.title("Marginal distribution of mean")
plt.show()

# Marginal distribution for log_precision
ind = torch.tensor(np.log(np.arange(prec_true*0.2,prec_true*2.,0.01)),dtype=torch.double)
plt.hist(np.log(gibbs_samples[1000:,1]), bins=20, density=True)
plt.plot(ind, torch.exp(laplace_log_prec.log_prob(ind)).detach())
plt.title("Marginal distribution of log-precision")
plt.show()
```

## Variational approximation

Next look at the variational approximation for this model, assuming factorized (mean-field) approximation. We switch back to numpy as we no longer need pytorch for optimization.

First define the model, approximation and optimize.

```python
import scipy.special

# Reuse the same data
if(isinstance(x, torch.Tensor)):
    x = x.numpy()

# Define the model again in numpy
mean_prior = stats.norm(prior_mu, prior_sigma)
prec_prior = stats.gamma(prior_alpha, scale=1.0/prior_beta)

def log_posterior(data, mu, prec):
    likelihood = stats.norm(loc=mu, scale=np.sqrt(1/prec))
    return np.sum(likelihood.logpdf(data)) + mean_prior.logpdf(mu) + prec_prior.logpdf(prec)

# Definte the approximation
# Assume factorized q(mu,sigma) = q(mu|m,s) q(sigma|a,b) (normal and gamma)
q_mu_param = [1.0, 0.5]
q_prec_param = [1.0, 1.0]
q_mu = stats.norm(q_mu_param[0], 1.0/np.sqrt(q_mu_param[1]))
q_prec = stats.gamma(q_prec_param[0], scale=1.0/q_prec_param[1]) # Irritating parameterization again

# Perform optimization; note how few iterations are enough here
for iter in range(10):
    ### Update q_mu
    # Compute required expectations
    E_prec = q_prec.mean()
    # If we were to implement the ELBO, we would also need the following expectation that is not
    # implemented in scipy.stats
    #E_log_prec = np.log(q_prec_param[1]) + scipy.special.digamma(q_prec_param[0])
    
    # Update the parameters
    q_mu_param[1] = 1.0/prior_sigma**2.0 + len(x)*E_prec
    q_mu_param[0] = (E_prec*np.sum(x) + 1.0/prior_sigma**2.0 * prior_mu) / q_mu_param[1]
    q_mu = stats.norm(q_mu_param[0], scale=1.0/np.sqrt(q_mu_param[1]))
    
    ### Update q_sigma
    # Compute required expectations
    E_mu = q_mu.mean()
    E_mu_var = q_mu.var()
    
    # Update the parameters
    q_prec_param[0] = prior_alpha + 0.5*len(x)
    q_prec_param[1] = prior_beta + 0.5*(len(x)*(E_mu**2.0 + E_mu_var) - 2.0*np.sum(x)*E_mu + np.sum(x**2.0))
    q_prec = stats.gamma(q_prec_param[0], scale=1/q_prec_param[1])
    
    # Here we could (and should) evaluate the objective -- try implementing it!
```

Then we can inspect the results. We see that the approximation quite closely matches the Laplace approximation -- for the mean we have almost the same distribution, and even though we cannot directly compare distributions for log-precision and precision we see they are very close.

```python
print("Parameters of q_mu: ", q_mu_param)
print("Parameters of q_prec: ", q_prec_param)

# Marginal distribution for mu
ind = np.arange(mu_true-1.5,mu_true+1.5,0.01)
plt.hist(gibbs_samples[1000:,0], bins=20, density=True)
plt.plot(ind, np.exp(q_mu.logpdf(ind)))
plt.title("Marginal distribution of mu")
plt.show()

# Marginal distribution for prec
ind = np.arange(prec_true*0.2,prec_true*2.,0.01)
plt.hist(gibbs_samples[1000:,1], bins=20, density=True)
plt.plot(ind, np.exp(q_prec.logpdf(ind)))
plt.title("Marginal distribution of precision")
plt.show()
```
