# Problem 1: Distributional approximations for the Gamma-Poisson model

Again start by defining the model (the same as last time)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
%matplotlib inline
plt.rcParams["figure.figsize"] = (12, 6)

N = 10
u_alpha = 5.
u_beta = 1.
v_alpha = 4.
v_beta = 4.

x = np.array([3, 4, 3, 9, 10, 3, 2, 3, 3, 2])

# Priors
u_prior = stats.gamma(u_alpha, scale=1./u_beta)
v_prior = stats.gamma(v_alpha, scale=1./v_beta)

def log_density(data, u, v):
    likelihood = stats.poisson(u*v)
    return np.sum(likelihood.logpmf(data)) + u_prior.logpdf(u) + v_prior.logpdf(v)

# ...and old Gibbs sampler for reference
def Gibbs(x, u, v, T):
    # Storage for samples
    samples = [u, v]
    sumx = sum(x)
    N = len(x)
    for t in range(T):
        # Sample u conditional on v and data
        log_lambda_term = sumx + u_alpha
        lambda_term = N*v + u_beta
        u = np.random.gamma(log_lambda_term, scale = 1.0 / lambda_term)
        
        # Sample v conditional on u and data
        log_lambda_term = sumx + v_alpha
        lambda_term = N*u + v_beta
        v = np.random.gamma(log_lambda_term, scale = 1.0 / lambda_term)

        samples = np.vstack([samples, [u, v]])
    return samples
```

### Laplace approximation

1. To find the MAP estimate I use naive grid-search, just to show that anything goes. For this model you could also  derive closed-form optimal solution based on the derivatives below, by solving the equations $\nabla \log p(X,u,v) = 0$.

2. The derivatives required for the Hessian can be computed as follows:

Starting with the joint density
$$
\log p(X,u,v) = \sum_i \left [ x_i \log u + x_i \log v - uv - \log x_i! \right ] \\
\quad\quad\quad - \alpha_u \log \beta_u - \log \Gamma(\alpha_u) + (\alpha_u - 1)\log u - \beta_u u \\
\quad\quad\quad - \alpha_v \log \beta_v - \log \Gamma(\alpha_v) + (\alpha_v - 1)\log v - \beta_v v
$$
we get
$$
\frac{d\log p(X,u,v)}{du} = \sum_i \left [x_i/u - v \right ] + (\alpha_u -1)/u - \beta_u
$$
and analogously for $v$.

Then we need to compute the second derivatives. We have
$$
\frac{d^2 \log p(X,u,v)}{dudu} = - \left ( \sum_i x_i + (\alpha_u-1) \right )/u^2
$$
with analogous expression for $v$, and finally the cross term is
$$
\frac{d^2 \log p(X,u,v)}{dudv} = - \sum_i 1 = -N.
$$

```python
# Find the MAP
grid = np.arange(0.05,5.,0.05)
best = -1e10
for i in range(len(grid)):
    for j in range(len(grid)):
        value = log_density(x, grid[i], grid[j])
        if(value > best):
            u = grid[i]
            v = grid[j]
            best = value
print("MAP for u: ", u)
print("MAP for v: ", v)

# Check that the derivatives are indeed approximately zero at the MAP
print("Derivative wrt u at MAP: ", np.round(np.sum(x/u - v) + (u_alpha - 1)/u - u_beta, decimals=5))
print("Derivative wrt v at MAP: ", np.round(np.sum(x/v - u) + (v_alpha - 1)/v - v_beta, decimals=5))

# Form the Hessian by collecting the negative second derivatives
H = np.zeros((2,2))
# THE FIRST RELEASE OF THE NOTEBOOK HAD A MISTAKE HERE!
H[0,0] = (np.sum(x) + u_alpha - 1) / u**2.0
H[1,1] = (np.sum(x) + v_alpha - 1) / v**2.0
H[0,1] = len(x)
H[1,0] = len(x)
print("Hessian: ")
print(H)
Cov = np.linalg.inv(H)
print("Covariance: ")
print(Cov)
```

Now we can inspect the results. Since the model is bivariate, the best way is to plot the contour of the approximation and compare it against the set of samples we got from Gibbs sampler (which here samples from the true posterior).

We see that Laplace approximation does capture some of the posterior mass around the mode, but because of the curved relationship between $u$ and $v$ the whole posterior cannot be modelled well with a single normal distribution. Instead, the tails are completely missed and we severely underestimate the variation.

```python
# Generate a grid over the 2D space
ugrid = np.arange(1,8,0.05)
vgrid = np.arange(0.2,2.0,0.05)
uset, vset = np.meshgrid(ugrid, vgrid)
uv = np.column_stack([uset.flat, vset.flat])

# Density values at the grid points
pdf_laplace = stats.multivariate_normal(np.array([u,v]), Cov).pdf(uv).reshape(uset.shape)


# Add some Gibbs samples on top to compare
gibbs_samples = Gibbs(x, 3.0, 0.5, 2500)
plt.plot(gibbs_samples[500:,0], gibbs_samples[500:,1], '.', alpha=0.2)

# Then plot the countours on top of the points
_ = plt.contour(uset, vset, pdf_laplace, levels=6)
```

### Let's then take a look at the variational approximation

Now we have an approximation that is a product of two gamma distributions. Based on the previous exercise we know the conditional distribution for $u$ is 
$$
p(u|v,X) = \text{Gamma}(\alpha_u + \sum_i x_i, \beta_u + Nv)
$$
and similarly for $v$.

To form the updates for $q(u)$ we simply replace all terms involving $v$ with the expectation over current $q(v)$. Since we only have $v$ appearing as it is, we only need $\mathbb{E}[v] = c/d$ and the updated parameters for $q(u)$ are hence $a=\alpha_u + \sum_i x_i$ and $b=\beta_u + N \frac{c}{d}$. The updates for $q(v)$ are analogous.

```python
# Some randomish initial values
a = 2.
b = 1.
c = 1.
d = 2.

q_u = stats.gamma(a, scale=1.0/b)
q_v = stats.gamma(c, scale=1.0/d)

for i in range(100):
    # Update q(u)
    E_v = q_v.mean()
    a = u_alpha + np.sum(x)
    b = u_beta + len(x)*c/d
    q_u = stats.gamma(a, scale=1.0/b)
    
    # Update q(v)
    E_u = q_u.mean()
    c = v_alpha + np.sum(x)
    d = v_beta + len(x)*a/b
    q_v = stats.gamma(c, scale=1.0/d)
    
    # Check that we are converging
    if np.mod(i,10)==0:
        print([i, E_u, E_v])

print("Mean estimate for u: ", q_u.mean())
print("Mean estimate for v: ", q_v.mean())
```

Even though our posterior approximation is now independent, we can still inspect the two-dimensional posterior in a similar manner as before.

We see that the approximation has similar problems as Laplace, by ignoring the tails of the curved distribution. In addition, we can see how the facorized mean-field approximation loses the correlation $u$ and $v$ have. However, the approximation better accounts for the shape of the posterior -- the true marginal distributions are closer to gamma distribution than normal and now the approximation captures that to some degree. In general, however, we did not gain much by switching to variational approximation since the posterior is unimodal and Laplace approximation hence is fairly okay.

The purpose of these exercises was to practice the derivations, but you also got a glimpse at how the approximations work in practice. Already for this very simple model we saw the the approximations capture some of the uncertainty and provides information about the true posterior, but they are clearly not modelling the true posterior in the same way as MCMC methods were. However, we also saw that it was quite a bit easier to investigate convergence and we got the results quite fast (though my grid search for MAP was slow).

```python
# density values at the grid points
pdf_VI = (q_u.pdf(uset)*q_v.pdf(vset)).reshape(uset.shape)

# Add some Gibbs samples on top to compare
plt.plot(gibbs_samples[500:,0], gibbs_samples[500:,1], '.', alpha=0.2)

# Then plot the countours on top of the points
_ = plt.contour(uset, vset, pdf_VI, levels=6)
```

# Problem 2: Variational approximation for mixture models

For the mixture model we start by sampling and plotting the data.

```python
import numpy as np
import scipy.stats as stats

np.random.seed(7879)

# Define the model
K = 3; N = 500
pi_true = np.array([0.35, 0.2, 0.45])
mu_true = np.array([0., -1.5, 1.5])
tau_true = np.array([1., 5., 5.])

# Sample data
N = 500
x = np.zeros((N,))
z_true = stats.multinomial(1,pi_true).rvs(N) # one-hot-encoding for easier implementation
colors = ['r','g','b']
plt.subplot(1,2,1)
for k in range(3):
    x[z_true[:,k]==1] = stats.norm(mu_true[k], 1./np.sqrt(tau_true[k])).rvs(sum(z_true[:,k]))
    plt.hist(x[z_true[:,k]==1], color=colors[k], bins=np.linspace(-4.,4.,50))
plt.subplot(1,2,2)
_ = plt.hist(x,bins=np.linspace(-4.,4.,50))
```

The mean-field approximation consists of normal distributions for the cluster means, gamma distributions for the cluster precisions, multinomial disrtributions for all cluster indicators, and a Dirichlet distribution for the cluster weights.

The implementation instantiates all of these with some initial values and then proceeds to make iterations where each term is updated at a time. The derivations for the updates are provided in a separate pdf for easier reading.

```python
from scipy.special import digamma

np.random.seed(78798)

# Control parameters
nIter = 100
K = 3
alpha = 1.
alpha0 = 0.5
beta0 = 0.5
tau0 = 0.5

# Define and initialize approximation
nu = np.full(K, N/K)              # Equal probability for all clusters in the beginning

xi = stats.dirichlet(nu).rvs(N)   # Each row of xi needs to sum up to one

a = np.full((K,), 1.)             # We don't care too much about initial precisions but avoid very large values
b = np.full((K,), 5.)

m = stats.norm(0., 1.).rvs((K,))  # Need to start with different ones to break symmetry
t = np.full((K,), 5.)             # Some uncertainty in beginning

# Store some parameters for monitoring convergence
history = np.zeros((nIter, 4*K))

for iter in range(nIter):
    # Update q(z)
    # Requires: Expectations of mu, mu^2, tau, log(tau) and log(pi)
    E_mu = m
    E_mu2 = m**2. + 1./t
    E_tau = a/b
    E_log_tau = digamma(a) - np.log(b)
    E_log_pi = digamma(nu) - digamma(np.sum(nu))
    
    for n in range(N):
        logprob = np.zeros(K)
        for k in range(K):
            logprob[k] = E_log_pi[k] + 0.5*E_log_tau[k] - 0.5*E_tau[k]*(x[n]**2. - 2.*x[n]*E_mu[k] + E_mu2[k])
        prob = np.exp(logprob - max(logprob))
        xi[n,:] = prob/np.sum(prob)
    
    # Update q(pi)
    # Requires: Expectations of I[z_n=k] = xi
    nu = np.sum(xi,0) + alpha
    if (iter % (nIter//10))==0:
        print('Expected cluster size parameters:',nu)
        
    # Update cluster parameters
    # Requires: Expectations of I[z_n=k], mu, mu^2, and tau. We can re-use the ones we computed
    #           at the beginning of the iteration, except for \tau that is updated here and needs to be
    #           re-evaluated for the mean term updates
    E_z = xi
    E_z_sum = np.sum(E_z,0)
    for k in range(K):
        # Precision
        a[k] = alpha0 + 0.5*E_z_sum[k]
        b[k] = beta0 + 0.5*np.sum(E_z[:,k]*(x**2. - 2.*x*E_mu[k] + E_mu2[k]))
        E_tau[k] = a[k]/b[k]   # Need to update this since we changed q(tau)

        # Mean
        t[k] = tau0 + E_z_sum[k]*E_tau[k]
        temp = E_tau[k]*np.sum(E_z[:,k]*x)
        m[k] = temp/t[k]

    history[iter,:] = np.concatenate([m, t, a, b])
```

We can now investigate how well the method works.

1. We see that the expected values for the cluster centroids, precisions and sizes approximately match the true values. That is, we were able to discover roughly the same structure. This is not at all guaranteed for data where the clusters overlap so much, and by changing the random seed you can probably find cases where the model converges to a different solution.

2. Since we now have explicit representation for the posterior approximation at every point, it is easy to study also the uncertainty about the parameters. The first two figures plot how both the mean estimates and their uncertainty behaves during the optimization, and the final one illustrates what we know about the cluster allocations at the end of the optimization. We see that the method converged in roughly 50 iterations.

```python
print("Means: ", E_mu)
print("Precisions: ", E_tau)
print("Weights: ", nu/np.sum(nu))
```

```python
plt.rcParams["figure.figsize"] = (16, 4)

# Illustration of convergence
colors = ['r-','g-','b-']
colors2 = ['r:','g:','b:']
plt.subplot(1,3,1)
for k in range(K):
    plt.plot(history[:,k], colors[k])
    # Two standard deviations around the mean
    plt.plot(history[:,k] + 2.*1./np.sqrt(history[:,k+K]), colors2[k])
    plt.plot(history[:,k] - 2.*1./np.sqrt(history[:,k+K]), colors2[k])
plt.title('Convergence of the cluster means')

plt.subplot(1,3,2)
for k in range(K):
    mean = history[:,k+2*K]/history[:,k+3*K]
    std = np.sqrt(history[:,k+2*K]/history[:,k+3*K]**2.0)
    plt.plot(mean, colors[k])
    plt.plot(mean + 2.*std, colors2[k])
    plt.plot(mean - 2.*std, colors2[k])
    
plt.title('Convergence of the precision parameters means')


# Uncertainty of cluster allocations
plt.subplot(1,3,3)
order = np.argsort(x)
for k in range(K):
    plt.plot(x[order], xi[order,k], colors[k])
_ = plt.title('Cluster probabilities for data points')
```
